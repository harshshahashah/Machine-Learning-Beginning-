```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Data.csv')
df1 = df['Internal Combustion Engine']
af1 = df1.plot(kind = 'bar')
af1.set_xticks(df.index)
af1.set_xlabel('Test ID')
af1.set_ylabel('Marks (out of 100)')
af1.set_title("Internal Combustion Engine")
```
