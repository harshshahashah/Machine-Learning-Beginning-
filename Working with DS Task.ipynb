Defining Set 1

```python
E = {1, 2, 4, 5, 8}
```

Defining Set 2

```python
N = {2, 3, 6, 7, 8}
```

Union

```python
M = E.union(N)
print("The union of E and N is :", M)
```

Intersection

```python
P = E.intersection(N)
print("The intersection of E and N is :", P)
```

Difference

```python
R = E.difference(N)
print("The difference of E and N is :", R)
```

Symmetric Difference

```python
S = E.symmetric_difference(N)
print("The symmetric difference of E and N is :", S)
```
